```python
!pip install samila

import matplotlib.pyplot as plt
from matplotlib import colors as mcolors
from samila import GenerativeImage, Projection
import random
import math

VALID_COLORS = list(dict(mcolors.BASE_COLORS, **mcolors.CSS4_COLORS).keys())
```

```python
# functions

def f1(x, y):
    result = random.uniform(-1,1) * x**2  - math.sin(y**2) + abs(y-x)
    return result
def f2(x, y):
    result = random.uniform(-1,1) * y**3 - math.cos(x**2) + 2*x
    return result

def f3(x, y):
    result = random.gauss(0, 1) * math.sin(y)+ (x + y) * random.uniform(-1, 1)
    return result
def f4(x, y):
    result = random.uniform(-1, 1) * y * x + math.cos(x ** 2) + random.gauss(0, 1)
    return result
```

```python
# basic
# default projection - rectlinear

g = GenerativeImage()
g.generate()
g.plot()
plt.show()
```

```python
# fully random
# Supported projections : RECTILINEAR, POLAR, AITOFF, HAMMER, LAMBERT and MOLLWEIDE

def plt_graph():
  func = random.choice([(f1, f2), (f3, f4)])
  projection = random.choice([Projection.POLAR, Projection.RECTILINEAR, Projection.AITOFF, Projection.HAMMER, Projection.LAMBERT, Projection.MOLLWEIDE])
  color = random.choice(VALID_COLORS)
  bgcolor = random.choice(['white', 'black'])
  g = GenerativeImage(*func)
  g.generate()
  g.plot(color=color, bgcolor=bgcolor, projection=projection)
  plt.show()
  print(f'Transformations: {func[0].__name__}, {func[1].__name__}')
  print(f'Projection: {projection}')
  print(f'Color: {color}')
  print(f'Seed: {g.seed}')



plt_graph()
```
